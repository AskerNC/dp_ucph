# Exercise 3 [L2]: Interpolation and continuous choice 

Consider the following Python code to create do linear interpolation:

```julia
f(x) =  (x.-3).^2

x_interp  = Array(LinRange(0,6,100))
f_true  = f(x_interp)
x_known = Array(0:1:6)

f_known = f(x_known)


interp      = LinearInterpolation(x_known,f_known);
f_linear_interp = interp(x_interp)
;
```

### 1. Plot the true function, the known points and the linearly interpolated function in a single figure.

```julia
plot(x_interp,f_true,label="True",
    xlabel=L"x", ylabel=L"f(x)", yguidefontrotation=-90)

plot!(x_interp,f_linear_interp,label="Interpolation")

plot!(x_known,f_known,seriestype = :scatter,label="Known values")

```

Consider the consumption-saving model with *continous* consumption choice and *infinite time* horizon:
$$\begin{aligned}
    V_{t}(W_{t}) & = \max_{c_{t}}\left\{ \sqrt{c_{t}}+\beta V_{t+1}(W_{t+1})\right\} \\
 & \text{s.t.}\\
W_{t+1} & = W_{t}-c_{t} \\
c_{t} & \leq  W_{t} , \,\forall t\\
c_{t} & \in  \mathbb{R}_+, \,\forall t
\end{aligned}$$
for $\beta=0.90$ and $W_{1}=10$.

### 2. Solve the model using *value function iteration*, *grid search*, and *linear interpolation*.

```julia
# Setup specifications in class. Defining a class is a neat way to access the parameters.
num_W =200
W = 10
par = (β=0.9,w=W,num_W= num_W,num_C=50,
        W_grid = Array(LinRange(0.0,W,num_W)), max_iter =1000,tol=10e-5)


# Solve the model
sol = ex3.solve_consumption_grid_search(par);
```

```julia
# define figure
function plot_solution(par,sol)
    p1  =plot(par.W_grid,sol.V,xlabel=L"W",ylabel=L"V")
    p2 = plot(par.W_grid,sol.C,xlabel=L"W",ylabel=L"C")
    plot(p1,p2,layout =(1,2),legend=false, yguidefontrotation=-90)
end

plot_solution(par,sol)
```
